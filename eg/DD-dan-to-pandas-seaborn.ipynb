# NB. this is a Python3 notebook!

The purpose of this notebook is to illustrate a raku-python hybrid workflow:

1. Use raku Dan for data acquisition & cleansing
2. Use raku Dan::Pandas to export to Python Pandas and save as .json (ie. DD-raku-dan-pandas.ipynb)
3. Use python Pandas to read json and python Matplotlib to create plots

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] = 8,4
import warnings
warnings.filterwarnings('ignore')
```

```python
stats = pd.read_json("DD-done.json")
```

```python
stats
```

```python
#Distribution:
visl =sns.distplot(stats['InternetUsers'],bins=100)
```

```python
#Boxplots:
vis2 = sns.boxplot(data=stats,x="IncomeGroup",y="BirthRate")
```

```python
#Seaborn Gallery
#https://seaborn.pydata.org/examples/index.html
```

```python
vis3 = sns.lmplot(data=stats,x='InternetUsers',y='BirthRate',fit_reg=False)
```

```python
vis3 = sns.lmplot(data=stats,x='InternetUsers',y='BirthRate',\
                  fit_reg=False,\
                  hue='IncomeGroup',
                  size=10)
```

```python
vis3 = sns.lmplot(data=stats,x='InternetUsers',y='BirthRate',\
                  fit_reg=False,                             \
                  hue='IncomeGroup',
                  size=10, \
                  scatter_kws={"s":200}
                 )
```
